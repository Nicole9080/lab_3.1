```python
#Задание_03_01_15
#Выполнила Манидарова Н.Ч.
#Группа ЦИБ-251

def gdp(c, i, g, ex, im):

    return c + i + g + ex - im

info_dict = {}
info_list = []

try:
    c = int(input("Конечное потребление: "))
    info_list.append(c)
    info_dict['c'] = c

    i = int(input("Валовое накопление капитала: "))
    info_list.append(i)
    info_dict['i'] = i

    g = int(input("Государственные расходы: "))
    info_list.append(g)
    info_dict['g'] = g

    ex = int(input("Экспорт: "))
    info_list.append(ex)
    info_dict['ex'] = ex

    im = int(input("Импорт: "))
    info_list.append(im)
    info_dict['im'] = im

    gdp_from_list = gdp(*info_list)

    gdp_from_dict = gdp(**info_dict)

    print(f"ВВП (через список) = {gdp_from_list}")
    print(f"ВВП (через словарь) = {gdp_from_dict}")

except ValueError:
    print("Ошибка: все значения должны быть целыми числами!")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```
